```python
import matplotlib.pyplot as plt
import numpy as np

data = np.arange(10)
print(data)
plt.figure()
plt.plot(data, np.sin(data),color='blue')

plt.figure()
plt.bar(data, np.sin(data))
plt.show()

data2 = np.random.rand(100,1)
plt.figure()
plt.hist(data2, color='red')
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,3,1)
ax2 = fig.add_subplot(2,3,2)
ax3 = fig.add_subplot(2,3,3)
ax4 = fig.add_subplot(2,3,4)
ax5 = fig.add_subplot(2,3,5)

ax1.plot(data, np.sin(data), 'k', data, np.sin(data), 'bo')
ax2.bar(data, np.sin(data))
ax3.hist(data2, color='red')
ax4.plot(data, np.cos(data), 'r--', data, np.cos(data), 'b*')
ax5.plot(data2, np.tan(data2), color='green')
plt.show()
```

```python
x = np.arange(0,2,0.0001)
y1 = np.sin(x)
y2 = np.cos(x)

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True)

ax1.fill_between(x, 0, y1)
ax1.set_ylabel('between y1 and 0')
ax2.fill_between(x, y1, 1)
ax2.set_ylabel('between y1 and 1')
ax3.fill_between(x, y1, y2)
ax3.set_ylabel('between y1 and y2')
ax3.set_xlabel('x')
plt.show()
```
